```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame.from_csv('atus_00003.csv')
df['weighted_sleep']=df.BLS_PCARE_SLEEP * df.WT06
df.iloc[1:40]
```

```python
state_group_sum = df.groupby(['YEAR', 'STATEFIP']).sum()

results = state_group_sum.weighted_sleep / state_group_sum.WT06
print(results[2005].mean())
print(results[2005])
```

```python
state_group = df.groupby(['YEAR', 'STATEFIP'])

results_state = state_group.BLS_PCARE_SLEEP.agg([np.mean, np.std])
print(results_state['mean'].mean())
print(results_state.loc[2005][0:5])

metro_group = df.groupby(['YEAR', 'METAREA'])
results_metro = metro_group.BLS_PCARE_SLEEP.agg([np.mean, np.std])
print(results_metro.loc[2005][0:5])

```
